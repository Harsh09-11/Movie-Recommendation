```python
moviemat= df.pivot_table(index="user_id",columns="title",values="rating")
```

```python
moviemat
```

```python
ratings.sort_values('num of rating',ascending = False).head()
```

```python
starwars_user_ratings=moviemat['Star Wars (1977)']
starwars_user_ratings.head()
```

```python
similar_to_startwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_startwars
```

```python
corr_starwars = pd.DataFrame(similar_to_startwars,columns=['Correlation'])
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending= False).head(10)
```

```python
corr_starwars= corr_starwars.join(ratings['num of rating'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of rating']>100].sort_values(by ='Correlation',ascending=False)
```

```python
def predict_movies(movie_name):
    movie_user_ratings = moviemat[movie_name]
    similar_to_movie = moviemat.corrwith(movie_user_ratings)
    
    corr_movie = pd.DataFrame(similar_to_movie,columns=['Correlation'])
    corr_movie.dropna(inplace = True)
    corr_movie= corr_movie.join(ratings['num of rating'])
    predictions= corr_movie[corr_movie['num of rating']>100].sort_values(by ='Correlation',ascending=False)
    
    return predictions
```

```python
predictions=predict_movies("Titanic (1997)")
predictions
```
